```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(10, 5)})
```

```python
df=pd.read_csv('world-happiness-report.csv')
df.head()
```

```python
df.info()
```

```python
from os import mkdir
try:
    mkdir("Plots")
except:
    pass
mypath = "Plots"
```

```python
def check_null(df):
    print(df.isna().sum())
        
check_null(df)
```

```python
df = df.fillna(df.mean())
check_null(df)
```

```python
year_group = df.groupby("year").sum()
year_group
```

```python
f, axes = plt.subplots(5,2, figsize = (20,20))
f.tight_layout(pad=8)
f.suptitle("Distribution Preprocessing")
cols = df.select_dtypes(exclude="object").columns
x_axis = 0
y_axis=0
for col in cols:
    sns.histplot(data=df, x=col, kde=True, ax=axes[x_axis, y_axis])
    axes[x_axis,y_axis].set_xlabel(col.capitalize())
    axes[x_axis,y_axis].set_ylabel("Count")
    axes[x_axis,y_axis].set_title(f"{col.capitalize()} Count")
    if y_axis == 1:
        y_axis = 0
        x_axis += 1
    else:
        y_axis += 1
plt.savefig(f"Plots/Distribution_Report.png")        
plt.show()
```

```python
df.groupby("Country name")["Freedom to make life choices"].mean().sort_values(ascending= False).head(10)
```

```python
f.suptitle("Social Support")

a=sns.barplot(x=year_group.index,y=year_group['Social support'].values)
a.tick_params(axis='x',rotation=90)
plt.title("Social Support")
f, plt.savefig(f"Plots/Social_support_year.png") 
```

```python
happiest_unhappiest = df[(df["Life Ladder"] > 7.5)|(df["Life Ladder"] < 3)]
happiest_unhappiest.sort_values('Life Ladder')
```

```python
f.suptitle("The Happiest and Unhappiest Countries in 2020")
plt.title("The Happiest and Unhappiest Countries in 2020")

sns.barplot(x = happiest_unhappiest["Life Ladder"], y = happiest_unhappiest["Country name"])

plt.savefig(f"Plots/Happiest_and_Unhappiest.png")
plt.show()
```

```python
f.suptitle("Joint Plot")
f, sns.jointplot(data=df, x="year", y="Social support")
plt.title("Joint plot")
plt.savefig(f"Plots/Jointplot.png")
```

```python
f, sns.jointplot(data=df, x="Life Ladder", y="Social support")
f.suptitle("Joint Plot2")
plt.title("Joint Plot2")
plt.savefig(f"Plots/Jointplot2.png")
```

```python
f.tight_layout(pad=8)
f.suptitle("Correlation")
sns.set(rc={'figure.figsize':(15, 10)})
plt.title('Correlation Matrix')
f, sns.heatmap(df.corr(), annot=True)
plt.savefig(f"Plots/Correlation.png")
```

Exporting Reports as PDF

```python
import os
from os import listdir, mkdir
```

```python
all_files = os.listdir("Plots")
reports = [f'Plots/{file}' for file in all_files]
print(reports)
```

```python
from fpdf import FPDF
```

```python
WIDTH=210
HEIGHT=297
```

```python
pdf=FPDF()
pdf.set_font("Arial","B",36)
pdf.add_page()
pdf.cell(WIDTH//2, HEIGHT//2,txt='REPORT',align='C')
for report in reports: 
    pdf.add_page()
    pdf.image(report,0,0,WIDTH)
pdf.output("Countries_Report")
```
